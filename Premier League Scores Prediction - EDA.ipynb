```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\User\Documents\matchesall2014-2020.csv",encoding = "latin-1")
df.drop('Unnamed: 0', axis = 1, inplace = True)
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df[['Home Goals Scored HT', 'Away Goals Scored HT']] = df['Half Time Score'].str.split('-', expand=True)
```

```python
df["Home Goals Scored HT"] = df[['Home Goals Scored HT']].astype("int")
df["Away Goals Scored HT"] = df[['Away Goals Scored HT']].astype("int")
```

```python
df.head()
```

```python
df.corr()
```

### For Home Team
From Correlation table above. It is observed that there is more positive correlation in Home Team Rating,Home Team On Target shots, Match excitement , Away Team goals conceeded, Home Goals Scored HT with Home Team Goals Scored
There is more negative correlation in Away team rating to Home Team Goals Scored.
We visualize Correlation between these features and target variable using Scatter Plots.

```python
sns.catplot(y="Home Team Goals Scored", x="Home Team Rating",data=df, aspect = 5)
plt.xlabel("Rating",fontsize=20)
plt.ylabel("Home Team Goals Scored",fontsize=20)
plt.show()
```

```python
sns.catplot(y="Home Team Goals Scored", x="Home Team On Target Shots",data=df, aspect = 5)
plt.xlabel("Home Team On Target Shots",fontsize=20)
plt.ylabel("Home Team Goals Scored",fontsize=20)
plt.show()
```

```python
sns.catplot(y="Home Team Goals Scored", x="Away Team Goals Conceeded",data=df, aspect = 5)
plt.xlabel("Away Team Goals Conceeded",fontsize=20)
plt.ylabel("Home Team Goals Scored",fontsize=20)
plt.show()
```

```python
sns.catplot(y="Home Team Goals Scored", x="Away Team Rating",data=df, aspect = 5)
plt.xlabel("Away Team Rating",fontsize=20)
plt.ylabel("Home Team Goals Scored",fontsize=20)
plt.show()
```

```python
sns.catplot(y="Home Team Goals Scored", x="Home Goals Scored HT",data=df, aspect = 5)
plt.xlabel("HT Score",fontsize=20)
plt.ylabel("Home Team Goals Scored",fontsize=20)
plt.show()
```

```python
features = df[['Home Team','Away Team','Away Team Rating','Away Team On Target Shots', 'Match Excitement', 'Home Team Goals Conceeded', 'Away Goals Scored HT', 'Home Team Rating', 'Home Team On Target Shots', 'Away Team Goals Conceeded', 'Home Goals Scored HT' ]]
features.head()
```

```python
features_one_hot = pd.get_dummies(features, columns = ['Home Team', 'Away Team'], drop_first = True)
features_one_hot.astype('float64').head()
```

```python
file = open('processed.csv', 'w', encoding = 'utf8')
features_one_hot.to_csv('processed.csv', index=False)
```
